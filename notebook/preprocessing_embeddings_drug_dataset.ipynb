```python
import pandas as pd

df = pd.read_csv("cleaned_drug_dataset.csv")

print(df.head())
```

```python
import sys
print(sys.executable)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

text = df["rag_text"].tolist()

embeddings = model.encode(text, show_progress_bar=True)
```

```python
import numpy as np

np.save(
    "drug_embeddings.npy",
    embeddings
)
```

```python
import faiss
import numpy as np

embeddings_dim = embeddings.shape[1]

index = faiss.IndexFlatL2(embeddings_dim)

index.add(np.array(embeddings))
```

```python
faiss.write_index(
    index,
    "drug_index.faiss"
)
```

```python
def search_drug(query, top_k=5):
    query_embedding = model.encode([query])

    distances, indices = index.search(
        np.array(query_embedding),
        top_k
    )

    results = df.iloc[indices[0]]

    return results[[
        "drug_name",
        "condition",
        "dosage_mg",
        "side_effects",
        "rag_text"
    ]]
```

```python
query = "diabetes metformin side effects"

results = search_drug(query, top_k=3)

print(results)
```
